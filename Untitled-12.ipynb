```python
def parsetxt(file):
    f=open(file)
    txt=f.read()
    txt=txt.replace("\n","")
    words= txt.split(",")
    f.close()
    return words

def urlgen(searchterm):
    urlterm = searchterm.replace(" ","+")
    url = "https://news.google.com/rss/search?q="+urlterm+"&hl=en-US&gl=US&ceid=US:en"
    return url

def getLinks(url):
    page = requests.get(url)
    doc = b(page.text,"xml")
    links = []
    for x in doc.find_all("link"):
        links.append(x.text)
    return links
    
def downloadArticle(links):
    for x in links:
        start = time.time()
        try:
            doc = requests.get(x,headers = headers)
            page = b(doc.text,"html.parser")
            filename = page.title.text
            filename = filename.replace(" ","_")+ ".html"
            elapsed = time.time() - start
            print(filename)
            print("ELAPSED: ",elapsed)
        except:
            print(x,"Error")
            f = open("error.csv",'a')
            f.writelines(x)
            f.close()
            print("elapsed: ",time.time()-start)

#        try:
#            saveArticle(page,page.title.text.replace(" ","_")+".html") 
#        except:
#            saveLog(x)


def checkRepo(df,repo):
    repo = repo.concat([df,repo])
    repo.drop_duplicates(keep = False)
    return()

def saveArticle(page, name):
    f = open(name+".html","w",encoding="utf-8")
    f.write(str(page.html))
    f.close

def saveLog(link):
    f = open("error.csv","a")
    f.write(link+"\n")
    f.close()
```
